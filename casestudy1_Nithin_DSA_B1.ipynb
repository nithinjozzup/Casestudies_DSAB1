# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1.1. Read the dataset to the python environment.

```python
cars='cars_data.csv'
df=pd.read_csv(cars)
```

# Printing the dataset

```python
df
```

# 2. Check for the null values present in the dataset.

```python
df.info()
j=0
for i in df:
    j+=df[i].isnull().sum()
print("\n\nThere are",j,"null values present in the dataset")
```

# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(10,30))
sns.set_theme(style='whitegrid')
sns.countplot(x='Buyer Gender',data=df)
plt.title('Male VS Female Buyers',size=20)
```

# Insights:
    1.Male buyers are above 5000 and Female buyers are below 5000

# Bar Graph with Color of car wise buyers

```python
plt.figure(figsize=(20,40))
sns.set_theme(style='darkgrid')
sns.countplot(x='Buyer Gender',hue='Color',data=df)
plt.title('Male VS Female Buyers based on color',size=20)
```

# Insights
    1.All the buyers are in the range of 200 to 300 based on each color.
    2.maximum buyers are for mauv color in male and orange color in female.
    3.minimum buyers are for fuscia color in male and green color in female.

# 4. Find the top 5 cars based on their sales price.

```python
df1=df.sort_values(by=['Sale Price'],ascending=False).head(5)
df1
```

```python
df2=df1[['Make','Model','Sale Price','Resell Price']]
df2
```

# Insight
    1.Volkswagen Jetta III is sold fot maximum price of 99997.65
    2.Two models of Lexus comes in top 5 cars based on their sales price ranking 3 and 5

# 5. Find the least 5 cars based on their Resell price.

```python
df3=df.sort_values(by=['Resell Price'],ascending=True).head(5)
df3
```

```python
df4=df3[['Make','Model','Sale Price','Resell Price']]
df4
```

# Insights
    1.Cheapest car retail price is for Chevrolet 1500 which has a resale value of 2000.84
