# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1.Reading the dataset

```python
cars='cars_data.csv'
df=pd.read_csv(cars)
```

# Printing the dataset

```python
df
```

# 2.Checking for null values

```python
df.info()
```

# 3 Bar Graph

```python
plt.figure(figsize=(20,10))
sns.set_theme(style='whitegrid')
sns.countplot(x='Buyer Gender',data=df)
plt.title('Male VS Female Buyers',size=20)
```

# Bar Graph with Color of car wise buyers

```python
plt.figure(figsize=(20,10))
sns.set_theme(style='darkgrid')
sns.countplot(x='Buyer Gender',hue='Color',data=df)
plt.title('Male VS Female Buyers based on color',size=20)
```

# 4. Top 5 cars based on sales price

```python
df1=df.sort_values(by=['Sale Price'],ascending=False).head(5)
df1
```
